# Task-2 Success of an upcoming Movie

```python
import pandas as pd 
import seaborn as sns
```

```python
df_movies = pd.read_csv("C:/Users/nites/Desktop/Movie_success_prediction_data\movie_success_rate.csv")
df_movies

```

### Scatterplot to visualize the relationship between RATINGS and VOTES of Movies.

```python
sns.scatterplot(data = df_movies, x = "Rating", y = "Votes")
```

### Regression plot to visualize the relationship between the 'runtime' of movies and 'ratings'.

```python
sns.regplot(data = df_movies, x = "Runtime (Minutes)", y = "Rating")
```

### Compare the performance of 'Directors' based on the 'ratings' of their movies.

```python
sns.violinplot(data = df_movies.nlargest(10, "Rating"), x = "Director", y = "Rating")
```
